iloc basically means index location.

```python
import pandas as pd
```

```python
bond = pd.read_csv("JB007.csv")
```

Below mentioned two commands i.e. bond.loc[15] and bond.iloc[15] will give same output at this point.

```python
bond.loc[15]
```

```python
bond.iloc[15]
```

```python
bond.iloc[[15,20]] # this wwill output data of row 15 and 20.
```

```python
bond.iloc[:4] # this will output data from row 0 to 3.
```

```python
bond.iloc[4:8] # this will output data from row 4 to 7.
```

```python
bond.iloc[20:] # it will print from 20th row till last row.
```

```python
bond.iloc[0:21:2] # it will print 0 to 20th row skipping 1 row in between each time.
```

```python
bond = pd.read_csv("JB007.csv", index_col="Film")
```

```python
bond.sort_index()
```

```python
bond.iloc[0] # Despite of index being set to "Film", it will work.
```

```python
bond.iloc[15] # it will display data of 15th row in dataset.
```

```python
bond.iloc[18:] # it will display from 18th row till last.
```

```python
bond.iloc[:5] # It will diaplay from 0th to 4th row.
```

```python
bond.iloc[[5,10,15,20]] # it will display data of row numbers mentioned in list i.e. 5, 10, 15 and 20.
```
