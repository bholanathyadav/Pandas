Rank method is used for ranking data in a particular column.

```python
import pandas as pd
```

```python
google = pd.read_csv("gsp.csv").dropna(how="all")
```

```python
google["High"] = google["High"].fillna(0).astype("int")
# Here, astype("int") method will convert all the data in "High" from float to int.
```

```python
google["High"].rank(ascending=False).astype(int)
```

```python
google["High Rank"] = google["High"].rank(ascending=False).astype(int)
```

```python
google.sort_values(by="High")
```

```python
google.sort_values(by="High", ascending=False)
```
