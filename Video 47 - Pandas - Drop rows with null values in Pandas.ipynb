```python
import pandas as pd
```

```python
google = pd.read_csv("gsp.csv")
```

```python
google.dropna() # it will drop all such rows in which any one of the column values are NaN.
```

```python
google.dropna(how="any") # in this case, any row in which even if a single column is NaN will be entirely deleted.
```

```python
google.dropna(how="all") # in this case, only those rows will be removed where all the column values are NaN.
```

```python
google.dropna(how="all", inplace=True) # this will reflect the changes in imported dataset as well.
```

```python
google.dropna(axis=1)
# In this case, columns will be taken into consideration instead of rows i.e. any column in which even if one value
# is NaN, that column will be dropped entirely.
```

```python
google.dropna(subset=["High"])
# it will take "High" 
```
