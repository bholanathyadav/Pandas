```python
import pandas as pd
```

```python
df = pd.read_csv("employee.csv", parse_dates=["DateofHire","DateofTermination"])
df["HispanicLatino"] = df['HispanicLatino'].astype('bool')
df['GenderID'] = df['GenderID'].astype('category')
```

```python
df.info()
```

```python
df['GenderID'] == 0
# here all the entries in 'GenderID' column will be compared to see if it is 0 (Zero). In case of a match, it will
# return "True", else "False".
```

```python
df[df["GenderID"] == 1] # It will output only those rows where "GenderID" is 1.
```

```python
df[df["Department"] == 'Sales'] # It will output only those rows where "Department" is "Sales".
```

Same thing can be done in a different way mentioned below:

```python
mask = df["Department"] == "IT/IS"
```

```python
df[mask] # Now, we will get only those rows where "Department" is "Production".
```

```python
df[df["Department"] != 'Admin Offices']
# We get all those rows with "Department" other than "Admin Offices".
```

```python
df[df["DOB"] <= "09/26/84"]
```
