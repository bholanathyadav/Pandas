```python
import pandas as pd
```

```python
google = pd.read_csv("gsp.csv")
```

```python
google.sort_values("High")
```

Previous statement has sorted the table in ascending order using values in "High" column. This function can be used on strings as well. It will sort the strings in A to Z order.

```python
google.sort_values("High", ascending=False)
```

When ascending=False, it will sort the table in increasing to decreasing order.

```python
google.sort_values("High", na_position="first") # Now, all NaN will appear first. There is no NaN in table as of now.
```

```python
google.sort_values("Open")
# Through this statement, data will be arranged by values under "Open" column with all NaN values occuring in last.
```

Currently there are no NaN values in our entire table, so it does not appear.
