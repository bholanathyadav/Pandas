```python
import pandas as pd
```

```python
bond = pd.read_csv("JB007.csv", index_col="Film")
```

```python
bond.sort_index(inplace=True)
```

For .query() method to work, it is mandatory that column names be string with no blank space.

```python
bond.columns = [column_name.replace(" ","_") for column_name in bond.columns]
```

```python
bond.head()
```

```python
bond.query('Bond_actor == "Sean Connery"')
```

```python
bond.query("Director == 'Terence Young'")
```

```python
bond.query("Bond_actor != 'Roger Moore'")
```

```python
bond.query('Bond_actor=="Roger Moore" and Director=="John Glen"')
```

```python
bond.query('Bond_actor=="Roger Moore" or Director=="John Glen"')
```

```python
bond.query('Bond_actor in ["Timothy Dalton","George Lazenby"]')
```

```python
bond.query("Bond_actor not in ['Sean Connery','Roger Moore']")
```
